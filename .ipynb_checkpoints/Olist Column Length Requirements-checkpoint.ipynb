```python
import os
import pandas as pd
import numpy as np
```

```python
# Get all the file names of datasets
datasets = [name for name in os.listdir() if (str(name)[-4:] == '.csv')]
```

```python
# Show the types and maximum character lengths of each column from each dataset
for file_name in datasets:
    dataset = pd.read_csv(file_name)
    print(file_name)
    for column in range(0,dataset.shape[1]):
        print(dataset.columns[column], '\n', 
              str(dataset.dtypes[column]), 
              max(dataset.iloc[:,column].apply(lambda x: len(str(x)))))
    print('\n')
```
